# **3. Analyze the data and generate the appropriate plot including its properties:**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

**a. Read each device’s date of release and sales. Show it using a line plot.**




```python
xlabel = df['Date of Release']
ylabel = df['Sales (USD)']

plt.figure(figsize=(12, 6))
plt.plot(xlabel,ylabel)
plt.title('Sales through Release')
plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.xticks(df['Date of Release'], rotation=45)
plt.show()
```

**b. Get the total sales and show line plot with the following style properties.**
-Line Style dotted and Line-color should be red

-Show legend at the lower right location.

-X label name = Year

-Y label name = Sales

-Add a circle marker.

-Line marker color as red

-Line width should be 3

```python
plt.figure(figsize=(12, 6))
plt.plot(df['Date of Release'], df['Sales (USD)'],
         linestyle='dotted', color='red', marker='o', markerfacecolor='red',
         linewidth=3, label='Total Sales')


xlabel = df['Date of Release']
ylabel = df['Sales (USD)']

plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Total Sales Over Time')
plt.xticks(rotation=45)
plt.legend(loc='lower right')
plt.tight_layout()
plt.grid(axis='y', linestyle='--',linewidth=1.00)
plt.show()

```

**c. Read all product sales data and show it using a multiline plot. Label them**

```python
plt.figure(figsize=(12, 6))
for device in df['Device'].unique():
    device_data = df[df['Device'] == device]
    plt.scatter(device_data['Date of Release'], device_data['Sales (USD)'], label=device)
plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.title('Sales of Different Technological Products Over Time')
plt.legend(loc='upper left', bbox_to_anchor=(1, 1)) #used to position the legend outside the main plot area
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--',linewidth=1.00)
plt.show()
```

**d. Read each brand sales data and show it using the bar chart. Label them**

```python
plt.figure(figsize=(12, 6))
xlabel = df['Brand']
ylabel = df['Sales (USD)']
plt.bar(xlabel, ylabel, color='g')
plt.xlabel('Brand')
plt.ylabel('Total Sales')
plt.title('Sales by Brand')
plt.show()
```

**f. Create a pie chart that shows each brand Country of Origin. Label them.**

```python
plt.figure(figsize=(12, 6))
brand_counts = df['Country of Origin'].value_counts()
plt.pie(brand_counts, labels=brand_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Country of Origin for Technological Product Brands')
plt.show()
```
