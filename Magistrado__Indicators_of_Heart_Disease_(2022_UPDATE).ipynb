```python
import pandas as pd
import matplotlib.pyplot as plt
```

**Which sex is more prevalence towards asthma?**

```python
df['Asthma'] = df['Asthma'].map({'Yes': 1, 'No': 0})

asthma_by_sex = df.groupby('Sex')['Asthma'].sum()

plt.figure(figsize=(12, 6))
plt.bar(asthma_by_sex.index, asthma_by_sex.values, color='hotpink')

plt.title('Asthma Prevalence by Sex')
plt.xlabel('Sex')
plt.ylabel('Number of Individuals with Asthma')
plt.xticks(rotation=0)

plt.show()
```

The graph shows that female are more prevalent to having asthma than males. This may be due to several reasons such as physical activity, immunities and environments.

```python
print(df['SleepTime'].unique())
print(df['BMI'].unique())

BMI_by_sleeptime = df.groupby('SleepTime')['BMI'].mean().reset_index()

plt.figure(figsize=(12, 6))
plt.plot(BMI_by_sleeptime['SleepTime'], BMI_by_sleeptime['BMI'], marker='o')

plt.title('Average BMI by Sleep Time')
plt.xlabel('Sleep Time (hours)')
plt.ylabel('Average BMI')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

 The graph is flunctuating and it is shown that the results are mix there are some people that have low BMI but sleeps long and high BMI but also sleeps long
