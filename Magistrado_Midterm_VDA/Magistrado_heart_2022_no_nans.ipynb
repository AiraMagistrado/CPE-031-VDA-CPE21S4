```python
import pandas as pd
import matplotlib.pyplot as plt
path="/content/drive/MyDrive/DataSet/heart_2022_no_nans.csv"
df= pd.read_csv(path, delimiter=',', header=0, encoding='utf-8') #Uses a comma as the column separator, Treat the first row as the header, use UTF-8 to support special characters
df
```

# **QUESTION**

1. What data visualization should be used to determine the relationship between sleep hours and BMI?

```python
plt.figure(figsize=(20, 10))
plt.scatter(df['SleepHours'],  df['BMI'], color='g')
plt.title("Relationship between Hours of Sleep and BMI")
plt.xlabel('Hours of Sleep')
plt.ylabel("BMI")
plt.show()
```

**Observation**: The scatter plot shows the relationship between hours of sleep and BMI.  The data indicates that the two variables do not appear to be clearly related.  Given that the data points are vertically clustered, people with a wide range of BMIs sleep for comparable amounts of time. There isn't any clear pattern or trend indicating that getting more sleep raises or lowers BMI.

2. What data visualization should be used to compare the number of alcohol drinkers by Sex?

```python
alcohol = df[df['AlcoholDrinkers'] == 'Yes']
counts = alcohol.groupby('Sex')['AlcoholDrinkers'].count()
plt.figure(figsize=(20, 10))
plt.bar(counts.index, counts.values, color='g')
plt.xlabel("Sex")
plt.ylabel("Number of Alcohol Drinkers")
plt.title("Alcohol Drinkers by Gender")
plt.show()
```

**Observation** :The bar chart shows the number of alcohol drinkers by sex. According to the data, the percentage of men and women who drink alcohol is nearly equal. There is a small difference, men drink alcohol at a slightly higher rate than women.
