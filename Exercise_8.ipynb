# **Format Strings fmt**

```python
#plotting x and y points
ypoints=np.array([3,8,1,10])

plt.plot(ypoints, '*:m') #marker | line | color
plt.show()
```

# **Marker Size**

```python
#plotting x and y points
ypoints=np.array([3,8,1,10])

plt.plot(ypoints, '*:m', ms=20) #marker | line | color
plt.show()
```

# **Marker Color**

 You can use the argument markeredgecolor or mec for

```python
#plotting x and y points
ypoints=np.array([3,8,1,10])

plt.plot(ypoints, '*:m', ms=20, mec='c') #marker | line | color
plt.show()
```

# **Marker Face Color or mfc**

```python
#plotting x and y points
ypoints=np.array([3,8,1,10])

plt.plot(ypoints, '*:m', ms=20, mfc='c') #marker | line | color
plt.show()
```

# **Matplot Labels and Title**

You can use the xlabel() and ylabel()

```python
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints, ypoints, 'm')  #plot() is used to draw points

plt.xlabel("Distance in km")
plt.ylabel("Speed in km/s")
plt.show()
```

```python
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints, ypoints, 'm') #plot() is used to draw points
plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("Speed in km/s")
plt.show()
```

# **Matplot Grid**

You can use grid() to add grid lines

```python
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints, ypoints, 'm') #plot() is used to draw points
plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("Speed in km/s")
plt.grid()
plt.show()
```

```python
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.plot(xpoints, ypoints, 'm') #plot() is used to draw points
plt.title("Distance Travelled Everyday")
plt.xlabel("Distance in km") #labels
plt.ylabel("Speed in km/s")
plt.grid(axis='x', linestyle='--',linewidth=0.50)
plt.show()
```

# **Show Multiple Plots**

```python
#plot 1
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])
plt.subplot(1,2,1)
plt.plot(xpoints, ypoints, 'hotpink')#plot() is used to draw points

#plot 2
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([10,20,30,40])
plt.subplot(1,2,2)
plt.plot(xpoints, ypoints, 'hotpink') #plot() is used to draw points

plt.grid(axis='x',linestyle='--',linewidth=0.50)
plt.show()
```

Matplot Scatter - The scatter function plots one dot for each observation

```python
#plot 1
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([3,8,1,10])

plt.scatter(xpoints,ypoints) #plot() is used to draw points

#plot 2
xpoints=np.array([1,2,6,8]) #plotting  x and y points
ypoints=np.array([10,20,30,40])

plt.scatter(xpoints,ypoints) #plot() is used to draw points

plt.show()
```

# **Matplotlib Bars**

- You can use the bar() to draw bars

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.array(['Student1','Student2','Student3'])
y=np.array([90,85,86])
plt.bar(x, y, color='hotpink')
plt.show()
```
